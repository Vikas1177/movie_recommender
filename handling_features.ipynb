```python
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv('data\movies.csv')
```

```python
df["genre_list"] = df["Genre"].str.split(",")
```

```python
genre_counts = Counter(genre for genres in df["genre_list"] for genre in genres)
```

```python
threshold = 40
def map_genres(genres):
    return [genre if genre_counts[genre] >= threshold else "Other" for genre in genres]

df["mapped_genres"] = df["genre_list"].apply(map_genres)
```

```python
all_genres = set(genre for genres in df["mapped_genres"] for genre in genres)
all_genres
```
