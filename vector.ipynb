```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_chroma import Chroma
```

```python
import pandas as pd

movies = pd.read_csv(r"data\movies.csv)
```

```python
movies
```

```python
movies["overview_with_code"].to_csv("overview_with_code.txt",sep = "\n",index = False,header = False)
```

```python
raw_documents = TextLoader("overview_with_code.txt").load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
movies_db = Chroma.from_documents(documents,embedding=embeddings)
```

```python
query = "A fiction movie with interesting concepts"
docs = movies_db.similarity_search(query, k = 10)
docs
```

```python
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 10,
) -> pd.DataFrame:
    recs = movies_db.similarity_search(query, k = 10)

    movies_list = []

    for i in range(0, len(recs)):
        movies_list += [int(recs[i].page_content.strip('"').split()[0])]

    return movies[movies["code"].isin(movies_list)]
```

```python
retrieve_semantic_recommendations("A movie about space")
```
