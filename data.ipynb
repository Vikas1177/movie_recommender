```python
data_new = data.drop(columns=['Meta_score','Gross'])
```

```python
data_new.isna().sum()
```

```python
data_new['IMDB_Rating'] = pd.to_numeric(data_new['IMDB_Rating'])
```

```python
data_new.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='No_of_Votes', y='IMDB_Rating', data=data_new, alpha=0.6, color='blue')

plt.title("No_of_Votes vs IMDB_Rating", fontsize=16)
plt.xlabel("No_of_Votes", fontsize=14)
plt.ylabel("IMDB_Rating", fontsize=14)

high_rated = data[data_new['IMDB_Rating'] > 8.0]
sns.scatterplot(x='No_of_Votes', y='IMDB_Rating', data=high_rated, color='red', label='High Rated (>8.0)', alpha=0.8)

sns.regplot(x='No_of_Votes', y='IMDB_Rating', data=data_new, scatter=False, color='green', label='Trendline')

plt.legend()

plt.show()
```

```python
data_new['len_of_Overview'] = data_new['Overview'].str.split().str.len()
```

```python
data_new
```

```python

genre_counts = data_new['len_of_Overview'].value_counts().reset_index()
genre_counts.columns = ['len_of_Overview', 'Count']

plt.figure(figsize=(12, 6))
sns.barplot(x='len_of_Overview', y='Count', data=genre_counts,hue='len_of_Overview', palette='viridis', legend=False)

plt.xticks(rotation=45, ha='right')
plt.xlabel("len_of_Overview", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.title("len_of_Overview Distribution", fontsize=14)
plt.show()
```
